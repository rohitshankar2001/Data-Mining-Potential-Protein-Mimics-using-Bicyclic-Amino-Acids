```python
#install all these packages before running the script
import os
import pandas as pd
import rdkit
from rdkit import Chem
from rdkit.Chem import Draw
from rdkit.Chem import AllChem
```

```python
#Locate folder
os.chdir(r"C:/Users/r25oh/Desktop/BicyclicClips")
#CSV file with thiols
thiolData = pd.read_csv('thiols.csv')
#CSV file with bromide scaffold
bromoData = pd.read_csv('bromideScaffolds.csv')
```

```python
#convert thiols in mol format
thiolCol = thiolData.iloc[:,0]
thiolArray = []
for i in thiolCol:
    mol = Chem.MolFromSmiles(i)
    Chem.AssignAtomChiralTagsFromStructure(mol)
    if mol is not None:
        thiolArray.append(mol)
        
#convert bromide scaffolds to mol format
bromoCol = bromoData.iloc[:,0]
bromoArray = []
for i in bromoCol:
    mol1 = Chem.MolFromSmiles(i)
    Chem.AssignAtomChiralTagsFromStructure(mol1)
    if mol1 is not None:
        bromoArray.append(mol1)
len(bromoArray)


```

```python
#made array with peptides 1-10 length and 3-10 length

#startingForSmallRing
alanine = Chem.MolFromSmiles('C[C@H](C(=O)O)N')
#startingForLargeRing
alanine2 = Chem.MolFromSmiles('C[C@H](C(=O)O)N')

rxnAddAla = AllChem.ReactionFromSmarts('[O:1]([C:2])>>[N:1]([C:2])[C@H:3]([C:4])[C:5](=[O:6])[O:7]')

#peptide arrays for each cycle
peptideSmallRingImages = []
peptideSmallRingSmiles = ['C[C@H](C(=O)O)N']

peptideLargeRingImages = []
peptideLargeRingSmiles = []

#will generate array of 9 peptide chain each one amino acid longer. This is for 1-10 ring
for x in range(9):
    
    test = rxnAddAla.RunReactants((alanine,))
    peptideSmallRingImages.append(test)

    alanine = Chem.MolToSmiles(test[0][0],True)
    peptideSmallRingSmiles.append(alanine)
    
    alanine = Chem.MolFromSmiles(alanine,True)

#will generate array of 9 peptide chain each one amino acid longer. This is for 3-10 ring
for x in range(9):
    
    
    test2 = rxnAddAla.RunReactants((alanine2,))
    peptideLargeRingImages.append(test2)

    alanine2 = Chem.MolToSmiles(test2[0][0],True)
    peptideLargeRingSmiles.append(alanine2)
    
    alanine2 = Chem.MolFromSmiles(alanine2,True)

#remove starting elements so length of 3 is first element in array
peptideLargeRingSmiles.pop(0)
peptideLargeRingImages.pop(0)


print(peptideLargeRingSmiles)    
#Draw.MolsToGridImage(peptideLargeRingImages[0][0])
```

```python
#Combines every possiblity of chain lengths with a thiol in the middle

rxnThiol = AllChem.ReactionFromSmarts('[O:2]([C:3](=[O:4])).[N:1] >> [N:1][C:3]=[O:4] ')

peptideSmallRingThiol = []

peptideCombineRingImages = []
peptideCombineRingSmiles = []

#Add thiol
for i in range(10):
    for u in range(len(thiolArray)):
    
        #reactsRings = (Chem.MolFromSmiles(peptideSmallRingSmiles[i]), Chem.MolFromSmiles('O=C(O)[C@H](N)CS'))
        
        reactsRings = (Chem.MolFromSmiles(peptideSmallRingSmiles[i]), thiolArray[u])
        productThiol = rxnThiol.RunReactants(reactsRings)

        #peptideSmallRingThiol.append(productThiol)

        productSmiles = Chem.MolToSmiles(productThiol[0][0],True)
        peptideSmallRingThiol.append(productSmiles)
    
#Combine so thiol in middle 
for x in range(len(peptideSmallRingThiol)):
    for y in range(8):
  
        reactsRings = (Chem.MolFromSmiles(peptideSmallRingThiol[x]),Chem.MolFromSmiles(peptideLargeRingSmiles[y]))
        products = rxnThiol.RunReactants(reactsRings)
        
        peptideCombineRingImages.append(products)
        
        test = Chem.MolToSmiles(products[0][0],True)
        peptideCombineRingSmiles.append(test)

        
#Draw.MolsToGridImage(peptideCombineRingImages[0][0])
#print(peptideCombineRingSmiles[80])


#Draw.MolsToGridImage(peptideSmallRingThiol[1][0])
print(peptideCombineRingSmiles[9:17])    
len(peptideCombineRingSmiles)

#products = rxn.RunReactants(reacts)
#Chem.MolToSmiles(products[0][0])
#len(peptideCombineRing)
#print(peptideCombineRing)

```

```python
#adding thiol groups at front and last places

peptideCombineRingEndThiol = []
peptideCombineRingEndThiolSmiles = []

PeptideCombineRingBothThiol = []
PeptideCombineRingBothThiolSmiles = []

#Thiol Placed At end thiol
rxn = AllChem.ReactionFromSmarts('[N:4].[O:1]([C:2](=[O:3])) >> [N:4][C:2](=[O:3])')

for x in range(len(peptideCombineRingSmiles)):
    for u in range(len(thiolArray)):
        #reacts2 = (Chem.MolFromSmiles('O=C(O)[C@H](N)CS'),Chem.MolFromSmiles('CC(N)C(=O)NC(C)C(=O)NC(C)C(=O)NC(C)C(=O)NC(C)C(=O)O'))
        reactsEndThiol = (thiolArray[u] ,Chem.MolFromSmiles(peptideCombineRingSmiles[x]))
        productsEndThiol = rxn.RunReactants(reactsEndThiol)
        
        peptideCombineRingEndThiol.append(productsEndThiol)

        productFirstThiolSmiles = Chem.MolToSmiles(productsEndThiol[0][0],True)
        peptideCombineRingEndThiolSmiles.append(productFirstThiolSmiles)


#Thiol placed at start
rxn2 = AllChem.ReactionFromSmarts('[O:0]([C:1](=[O:2])).[N:3] >> [N:3][C:1]=[O:2]')

#rxn2 = AllChem.ReactionFromSmarts('[C:0].[O:3] >> [O:3]')
#rxn2 = AllChem.ReactionFromSmarts('[O:0](=[O:1]).[N:2]([C:3]) >> [O:0](=[O:1])[N:2]([C:2])')
#rxn2 = AllChem.ReactionFromSmarts('[N:1].[O:1]([C:2](=[O:3])) >> [N:1][C:2](=[O:3])')

for x in range(len(peptideCombineRingEndThiolSmiles)):
    for u in range(len(thiolArray)):
        #reactsStartThiol = (Chem.MolFromSmiles('CC(N)C(=O)NC(S)C(=O)O'), thiolArray[u])
        #reactsStartThiol = (Chem.MolFromSmiles(peptideComebineRingReplacedN[x]), thiolArray[u])
        reactsStartThiol = (thiolArray[u],Chem.MolFromSmiles(peptideCombineRingEndThiolSmiles[x]))
        
        productStartThiol = rxn2.RunReactants(reactsStartThiol)

        PeptideCombineRingBothThiol.append(productStartThiol)
        productSecondThiolSmiles = Chem.MolToSmiles(productStartThiol[0][0],True)
        PeptideCombineRingBothThiolSmiles.append(productSecondThiolSmiles)
        
print(len(PeptideCombineRingBothThiolSmiles))
print(PeptideCombineRingBothThiolSmiles[0:36])
```

```python
#Add bromide to S 

bromideOneAdded = []

bromideReplaceBr = []
bromideReplaceBr2 = []

bromideReplaceS = []
bromideReplaceS2 = []

bromideTwoAdded = []
bromideAllAdded = []

bromideAllAddedReplaced1 = []
bromideAllAddedReplaced2 = []


for x in range(len(bromoArray)):
    replace =  Chem.MolToSmiles(bromoArray[x])
    replaced = replace.replace("Br" , "F" , 1)
    bromideReplaceBr.append(replaced)
    
    
for x in range(len(bromideReplaceBr)):
    replace =  bromideReplaceBr[x]
    replaced = replace.replace("Br" , "Cl" , 1)
    bromideReplaceBr2.append(replaced)
    
#print(bromideReplaceBr2)   
    
for x in range(len(PeptideCombineRingBothThiolSmiles)):
    replace = PeptideCombineRingBothThiolSmiles[x]
    replaced = replace.replace("CS)C" , "CI)C", 1 )
    bromideReplaceS.append(replaced)
#print(bromideReplaceS)
    
for x in range(len(bromideReplaceS)):
    replace = bromideReplaceS[x]
    replaced = replace.replace("CS)C" , "CP)C", 1 )
    bromideReplaceS2.append(replaced)
#print(bromideReplaceS2)    

rxnBromo = AllChem.ReactionFromSmarts('[Br:0].[P:2]([C:3]) >> [P:0]([C:3])')
#rxnBromo = AllChem.ReactionFromSmarts('[Br:0][C:1][c:2][c:3][c:4][C:5][F:6].[I:7][S:8] >> [Br:0][C:1][c:2][c:3][c:4][C:5][F:6][I:7][S:8] ')
#rxnBromo = AllChem.ReactionFromSmarts('[Br:0][C:1][c:2][c:3][c:4]([C:5]([F:6]))[c:7].[I:10]>> [Br:0][C:1][c:2][c:3][c:4]([C:5]([F:6]))[c:7][I:10]')

#rxnBromo = AllChem.ReactionFromSmarts('[F:0][Br:1].[I:2]([C:3])[S:4]([C:5]) >> [S:0]([C:3])[S:1]([C:5])')

for x in range(len(bromideReplaceS2)):
    for u in range(len(bromoArray)):
        bromoReactants = (Chem.MolFromSmiles(bromideReplaceBr2[u]),Chem.MolFromSmiles(bromideReplaceS2[x]))
        bromoProduct = rxnBromo.RunReactants(bromoReactants)
                
        bromoSmile = Chem.MolToSmiles(bromoProduct[0][0],True)
        bromideOneAdded.append(bromoSmile)
        
#print(bromideOneAdded)
#print(bromideReplaceS)

rxnBromo2 = AllChem.ReactionFromSmarts('([F:0].[I:1]([C:2]))>>[I:0]([C:2])')

#bromoReac = (Chem.MolFromSmiles(bromideOneAdded[0]),)
#bromoProd = rxnBromo2.RunReactants(bromoReac)
#bromoSm2 = Chem.MolToSmiles(bromoProd[0][0],True)

#print(bromoSm2)

for x in range(len(bromideOneAdded)):
    bromoReactants2 = (Chem.MolFromSmiles(bromideOneAdded[x]),)
    bromoProduct2 = rxnBromo2.RunReactants(bromoReactants2)
        
    bromoSmile2 = Chem.MolToSmiles(bromoProduct2[0][0],True)
    bromideTwoAdded.append(bromoSmile2)
        
#print(bromideTwoAdded)

rxnBromo3 = AllChem.ReactionFromSmarts('([Cl:0].[S:1]([C:2]))>>[S:0]([C:2])')

for x in range(len(bromideTwoAdded)):
    
    bromoReactants3 = (Chem.MolFromSmiles(bromideTwoAdded[x]),)
    bromoProduct3 = rxnBromo3.RunReactants(bromoReactants3)
    bromoSmile3 = Chem.MolToSmiles(bromoProduct3[0][0],True)
    bromideAllAdded.append(bromoSmile3)
    
#print(bromideAllAdded)

rxnAddS = AllChem.ReactionFromSmarts('[P:2]([C:3]) >> [S:2]([C:3])')
for x in range(len(bromideAllAdded)):
    
    bromoReactants4 = (Chem.MolFromSmiles(bromideAllAdded[x]),)
    bromoProduct4 = rxnAddS.RunReactants(bromoReactants4)
    bromoSmile4 = Chem.MolToSmiles(bromoProduct4[0][0],True)
    bromideAllAddedReplaced1.append(bromoSmile4)   
    
#print(bromideAllAddedReplaced1)

rxnAddS2 = AllChem.ReactionFromSmarts('[I:2]([C:3]) >> [S:2]([C:3])')
for x in range(len(bromideAllAddedReplaced1)):
    
    bromoReactants5 = (Chem.MolFromSmiles(bromideAllAddedReplaced1[x]),)
    bromoProduct5 = rxnAddS2.RunReactants(bromoReactants5)
    bromoSmile5 = Chem.MolToSmiles(bromoProduct5[0][0],True)
    bromideAllAddedReplaced2.append(bromoSmile5)   
    
print(bromideAllAddedReplaced2)
    
```

```python
def generateconformations(m):
    #add hydrogen atoms
    m = Chem.AddHs(m)
    #generate 5 conformers
    ids=AllChem.EmbedMultipleConfs(m, numConfs=5)
    for id in ids:
        #geometrically optimze each conformer
        AllChem.UFFOptimizeMolecule(m, confId=id)
    return m
```

```python
#Change the C terminus(COOH) to CONH2 and NH2 terminus to NH-CO-CH3
nTermChanged = []
bothTermChanged = []

#reaction to change NH2 to NH-CO-CH3
rxnNTerm = AllChem.ReactionFromSmarts('[NH2:0] >> [NH:0][C:1](=[O:2])[CH3:3]')

#reaction to change COOH to CONH2
rxnCTerm = AllChem.ReactionFromSmarts('[OH:1] >> [NH2:1]')


#apply reaction n term change
for x in range(len(bromideAllAddedReplaced2)):
    nTermChangeReactants = (Chem.MolFromSmiles(bromideAllAddedReplaced2[x]),)
    nTermProduct = rxnNTerm.RunReactants(nTermChangeReactants)
    nTermSmiles = Chem.MolToSmiles(nTermProduct[0][0],True)
    nTermChanged.append(nTermSmiles)
    
for x in range(len(nTermChanged)):
    bothTermChangeReactants = (Chem.MolFromSmiles(nTermChanged[x]),)
    bothTermProduct = rxnCTerm.RunReactants(bothTermChangeReactants)
    bothTermSmiles = Chem.MolToSmiles(bothTermProduct[0][0],True)
    bothTermChanged.append(bothTermSmiles)

print(bothTermChanged)

#t= Chem.MolFromSmiles('C[C@H]1NC(=O)[C@H](N)CSCc2cc3cc(c2)CSC[C@@H](NC1=O)C(=O)N[C@H](C)C(=O)N[C@H](C)C(=O)N[C@H](C)C(=O)N[C@@H](C(=O)O)CSC3')
#test = rxnNTerm.RunReactants((t,))     

#Chem.MolToSmiles(test[0][0],True)
```

```python
#save each conformer in mol format in your disk

for i in range(len(bromideAllAddedReplaced2)):
    x = Chem.MolFromSmiles(bromideAllAddedReplaced2[i])
    #generate conformers
    m = generateconformations(x)
    #get ids of conformers
    confIds = [x.GetId() for x in m.GetConformers()]
    #use the 1st one to save as 'mol' file. Sometimes there were failed cases in generating conformers, that's why we need to prepare more than 1 conformer. Here I prepare 5.
    l = Chem.MolToMolBlock(m, confId=1)
    #your folder to save these conformers
    dire = r'C:\Users\r25oh\Desktop\BicyclicClips'
    if os.path.isdir(dire) == False:
        os.mkdir(dire)
    #save coordinates to a mol file
    molfile = open("Tye_3R//"+"3R"+str(i)+".mol","w")
    molfile.write(l)
    molfile.close()
```

```python
#save each conformer WITH CHANGED TERMINUS in mol format in your disk

for i in range(len(bothTermChanged)):
    x = Chem.MolFromSmiles(bothTermChanged[i])
    #generate conformers
    m = generateconformations(x)
    #get ids of conformers
    confIds = [x.GetId() for x in m.GetConformers()]
    #use the 1st one to save as 'mol' file. Sometimes there were failed cases in generating conformers, that's why we need to prepare more than 1 conformer. Here I prepare 5.
    l = Chem.MolToMolBlock(m, confId=1)
    #your folder to save these conformers
    dire = r'C:\Users\r25oh\Desktop\BicyclicClips'
    if os.path.isdir(dire) == False:
        os.mkdir(dire)
    #save coordinates to a mol file
    molfile = open("SaveLocation//"+"3R"+str(i)+".mol","w")
    molfile.write(l)
    molfile.close()
```
